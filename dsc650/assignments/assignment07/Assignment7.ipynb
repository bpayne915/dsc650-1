```python
df = pd.read_parquet('~/dsc650/dsc650/assignments/assignment03/results/routes.parquet')
```

```python
def key(row):
    try:
        airline = row.airline.get('iata')
        src_airport = row.src_airport['iata']
        dst_airport = row.dst_airport['iata']
        key = f'{src_airport}{dst_airport}{airline}'
    except:
        key = 'BBBBBBBB'
    return(key)
```

```python
df['key'] = df.apply(key, axis=1)
```

```python
df.head()
```

```python
def partition(row):
    key_first_letter = row.key[0]
    if key_first_letter in ['C','D']:
        key_first_letter = 'C-D'
    if key_first_letter in ['E','F']:
        key_first_letter = 'E-F'
    if key_first_letter in ['G','H']:
        key_first_letter = 'G-H'
    if key_first_letter in ['I','J']:
        key_first_letter = 'I-J'
    if key_first_letter in ['K','L']:
        key_first_letter = 'K-L'
    if key_first_letter in ['O','P']:
        key_first_char = 'O-P'
    if key_first_letter in ['Q','R']:
        key_first_letter = 'Q-R'
    if key_first_letter in ['S','T']:
        key_first_char = 'S-T'
    if key_first_letter in ['W','X']:
        key_first_letter = 'W-X'
    if key_first_letter in ['Y','Z']:
        key_first_letter = 'Y-Z'
    return(key_first_letter)
```

```python
df['kv_key'] = df.apply(partition, axis=1)
```

```python
df.head()
```

```python
df.to_parquet('./results/kv/',partition_cols=['kv_key'])
```

```python
import hashlib

def hash_key(key):
    m = hashlib.sha256()
    m.update(str(key).encode('utf-8'))
    return m.hexdigest()
```

```python
def make_hash_key(row):
    key = row.key
    return(hash_key(key)[0])
```

```python
df['hash_key'] = df.apply(make_hash_key, axis=1)
```

```python
df
```

```python
df.to_parquet('./results/hash/',partition_cols=['hash_key'])
```

```python
location_data = [
    {
        "location": "west",
        "city": "The Dalles, Oregon",
        "latitude": 45.5945645,
        "longitude": -121.1786823
    },
    {
        "location": "central",
        "city": "Papillion, NE",
        "latitude": 41.1544433,
        "longitude": -96.0422378
    },
    {
        "location": "east",
        "city": "Loudoun County, Virginia",
        "latitude": 39.08344,
        "longitude": -77.6497145
    }
]
```

```python
for location in location_data:
    location['geohash'] = pygeohash.encode(location['latitude'], location['longitude'])
    
location_data
```

```python
def location_search(latitude, longitude, location_data):
    try:
        search_hash = pygeohash.encode(latitude, longitude)
        closest_location_distance = math.inf
        for location in location_data:
            distance = pygeohash.geohash_haversine_distance(search_hash, location['geohash'])
            if distance < closest_location_distance:
                closest_location = location
                closest_location_distance = distance
        return(closest_location['location'])
    except Exception as e:
        return('central')
```

```python
def make_key(row):
    try:
        latitude = row.src_airport['latitude']
        longitude = row.src_airport['longitude']
    except:
        latitude = None
        longitude = None
    key = location_search(latitude, longitude, location_data)
    return(key)
```

```python
location_search(31.949288, 55.2739, location_data)
```

```python
location_search(65.794235,-92.521243, location_data)
```

```python
df['location'] = df.apply(make_key, axis=1)
```

```python
df.to_parquet('./results/geo/', partition_cols=['location'])
```

```python
def balance_partitions(keys, num_partitions):
    partition_size, remainder = divmod(len(keys), num_partitions) 
    partitions = []
    for i in range(num_partitions):
        start_index = i * partition_size + min(i, remainder)
        finish_index = (i + 1) * partition_size + min(i + 1, remainder)
        partition = keys[start_index : finish_index]
        partitions.append(partition)
    return partitions
```

```python
list_key = [1,3,2,4,6,5,7,8, 9]
```

```python
for i in range(1,len(list_key)+1):
    print(f'{i} partitions: {balance_partitions(list_key, i)}')
```
